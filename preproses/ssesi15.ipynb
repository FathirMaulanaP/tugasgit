BEBERAPA SKENARIO DAN APA YANG DAPAT DIANALISIS DARI DATASET TERSEBUT
| Pertanyaan                                         | Analisis yang Dilakukan                                               |
| -------------------------------------------------- | --------------------------------------------------------------------- |
| **Merek mana yang paling sering muncul?**          | Distribusi `brand_name`                                               |
| **Berapa rata-rata harga per merek?**              | `groupby` brand lalu `mean(price)`                                    |
| **Apakah 5G memengaruhi harga?**                   | Perbandingan `price` antara `has_5g=True` dan `False`                 |
| **Hubungan RAM dan Harga?**                        | Korelasi `ram_capacity` vs `price`                                    |
| **Fitur mana yang paling umum?**                   | Persentase fitur seperti `has_nfc`, `has_ir_blaster`, `fast_charging` |
| **Apakah ada korelasi rating dengan spesifikasi?** | Korelasi `rating` dengan fitur-fitur teknis                           |
