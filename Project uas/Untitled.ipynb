```python
from sklearn.linear_model import LinearRegression
df = pd.read_csv('Hours and Scores.csv')

X = df[['Hours']]
y = df[['Scores']]

model = LinearRegression()
model.fit(X,y)

plt.scatter(X, y, color='red' ,marker='o' ,label='Data asli')
plt.plot(X ,model.predict(X), color='blue' ,label='Garis Regresi')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Regresi Linear : Jam belajar dan Nilai')
plt.legend()
plt.grid(True)
plt.show()

nilai_prediksi = model.predict([[7]])
print(f"Prediksi nilai ujian jika 7 jam belajar: {nilai_prediksi[0][0]:,.2f}")
print(f"Koefisien setiap 1 jam belajar: {model.coef_[0][0]:,.2f}")
```
